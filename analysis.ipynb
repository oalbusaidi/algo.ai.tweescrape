```python
import re
import sqlite3
import pandas as pd

connection = sqlite3.connect('scraped_data.db')

tweets_df = pd.read_sql_query("SELECT * from twitter_tweet", connection)
user_df = pd.read_sql_query("SELECT * from twitter_user", connection)
```

```python
def clean_text (text) :
    text = re.sub(r'/', '', text)
    text = re.sub(r'\:', '', text)
    text = re.sub(r'\.', '', text)
    text = re.sub(r'\\', '', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[a-zA-Z]', '', text)
    text = re.sub(r'http[s]?://\S+', '', text)
    return text
```

```python
tweets_df['cleanedContent'] = tweets_df['rawContent'].apply(clean_text)
```

```python
tweets_df = tweets_df[tweets_df['lang']!= 'en']
```

```python
tweets_selected = tweets_df[['userid', 'rawContent', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'viewCount', 'date', 'cleanedContent']]
```

```python
tweets_selected[tweets_selected.duplicated(subset=['rawContent'])]
```

```python
tweets_no_duplicates = tweets_selected.drop_duplicates(subset=['cleanedContent'])
tweets_no_duplicates
```

```python
tweets_no_duplicates.loc[tweets_no_duplicates['viewCount'].isna(), 'viewCount'] = 0
```
